```python
# Select relevant columns and create combined text
merged_df['combined_text'] = merged_df.apply(lambda row: f"Title: {row['Title']} Score: {row['review/score']} Text: {row['review/text']}", axis=1)

# Select the necessary columns
ratings = merged_df[['User_id', 'Title', 'combined_text', 'review/score']]
```

```python
ratings.to_csv("user_movie_metadata.csv", index=False)
```

```python
from langchain_community.document_loaders.csv_loader import CSVLoader
from transformers import TFAutoModel

# Specify the encoding when creating the CSVLoader
loader = CSVLoader(file_path="user_movie_metadata.csv", encoding="utf-8")
data = loader.load()

from langchain_chroma import Chroma
from langchain_community.embeddings.sentence_transformer import (
    SentenceTransformerEmbeddings,
)

fine_tuned_model = TFAutoModel.from_pretrained('fine_tuned_all_mini_lm_l6_v2')

# Create the open-source embedding function
embedding_function = SentenceTransformerEmbeddings(fine_tuned_model)

folder_db = "db"
# Load it into Chroma
db = Chroma.from_documents(data, embedding_function, persist_directory=folder_db)
```

```python
persist_directory = "db"
embedding = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2")

# Load it into Chroma
vectordb = Chroma(persist_directory=persist_directory, embedding_function=embedding)

def get_combined_text_by_user_id(user_id):
    # Locate the combined text for the given user_id
    user_text = ratings.loc[ratings['User_id'] == user_id, 'combined_text']
    if not user_text.empty:
        return user_text.values[0]
    else:
        raise ValueError(f"No combined text found for user_id: {user_id}")

# Example usage
user_id = 'A36TPZSH8LBT1'
combined_text = get_combined_text_by_user_id(user_id)

# Perform similarity search using the combined text
similarity_questions = vectordb.similarity_search(combined_text, k=10)
print(similarity_questions)
```
