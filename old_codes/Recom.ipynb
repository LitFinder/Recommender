```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Combine relevant text features for vectorization
book_df['combined_features'] = book_df['description'] + ' ' + book_df['categories'] 

tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(book_df['combined_features'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import heapq

def compute_top_n_cosine_similarity(tfidf_matrix, batch_size=1000, top_n=10):
    num_docs = tfidf_matrix.shape[0]
    top_n_similarities = []

    for start_idx in range(0, num_docs, batch_size):
        end_idx = min(start_idx + batch_size, num_docs)
        batch = tfidf_matrix[start_idx:end_idx]
        
        # Compute similarity for the batch
        sim_batch = cosine_similarity(batch, tfidf_matrix)
        
        for i in range(sim_batch.shape[0]):
            row_similarities = sim_batch[i]
            # Get the top-N indices and their similarities
            if len(row_similarities) > top_n:
                top_indices = heapq.nlargest(top_n, range(len(row_similarities)), row_similarities.take)
                top_n_similarities.append((start_idx + i, [(idx, row_similarities[idx]) for idx in top_indices]))
            else:
                top_indices = range(len(row_similarities))
                top_n_similarities.append((start_idx + i, [(idx, row_similarities[idx]) for idx in top_indices]))

    return top_n_similarities

# Compute top-N cosine similarities
top_n_similarities = compute_top_n_cosine_similarity(tfidf_matrix, batch_size=1000, top_n=10)
```

```python
# Function to get recommendations
def get_content_based_recommendations(title, top_n_similarities, indices, book_df, top_n=10):
    idx = indices[title]
    similar_indices = [x[0] for x in top_n_similarities[idx][1]]
    similar_books = book_df.iloc[similar_indices]['Title']
    return similar_books

# Build indices for quick lookup
indices = pd.Series(book_df.index, index=book_df['Title']).drop_duplicates()

# Example usage
print(get_content_based_recommendations('The Battleship Bismarck', top_n_similarities, indices, book_df))
```

```python
from sklearn.model_selection import train_test_split

# Encode user and book IDs
user_ids = rating_df['User_id'].unique().tolist()
book_ids = rating_df['Id'].unique().tolist()

user_to_index = {x: i for i, x in enumerate(user_ids)}
book_to_index = {x: i for i, x in enumerate(book_ids)}

rating_df['user'] = rating_df['User_id'].map(user_to_index)
rating_df['book'] = rating_df['Id'].map(book_to_index)

# Train-test split
train, test = train_test_split(rating_df, test_size=0.2, random_state=42)
```

```python
num_users = len(user_ids)
num_books = len(book_ids)
embedding_size = 50

# Define input layers
user_input = tf.keras.layers.Input(shape=(1,))
book_input = tf.keras.layers.Input(shape=(1,))

# Define embedding layers
user_embedding = tf.keras.layers.Embedding(num_users, embedding_size)(user_input)
book_embedding = tf.keras.layers.Embedding(num_books, embedding_size)(book_input)

# Flatten the embeddings
user_vector = tf.keras.layers.Flatten()(user_embedding)
book_vector = tf.keras.layers.Flatten()(book_embedding)

# Compute dot product of user and book embeddings
dot_product = tf.keras.layers.Dot(axes=1)([user_vector, book_vector])

# Define model
model = tf.keras.Model(inputs=[user_input, book_input], outputs=dot_product)
model.compile(optimizer='adam', loss='mse')

# Print model summary
model.summary()
```

```python
# Prepare training data
train_user_data = train['user'].values
train_book_data = train['book'].values
train_ratings = train['review/score'].values

# Train the model
history = model.fit([train_user_data, train_book_data], train_ratings, epochs=10, verbose=1)
```

```python
import numpy as np

def get_collaborative_recommendations(user_id, model=model, top_n=10):
    user_idx = user_to_index[user_id]
    book_ids = np.array(merged_df['Id'].unique())
    book_indices = np.array([book_to_index[id] for id in book_ids])
    
    user_array = np.full(len(book_indices), user_idx)
    predictions = model.predict([user_array, book_indices])
    
    top_indices = np.argsort(predictions[:, 0])[::-1]
    
    unique_titles = set()
    recommended_books = []

    for idx in top_indices:
        book_id = book_ids[idx]
        book = merged_df[merged_df['Id'] == book_id].iloc[0]
        if book['Title'] not in unique_titles:
            unique_titles.add(book['Title'])
            recommended_books.append(book['Title'])
        if len(recommended_books) == top_n:
            break

    return recommended_books

# Example usage
print(get_collaborative_recommendations('ANX3DDV12ZRRU'))
```
