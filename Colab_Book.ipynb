Cosine Similarity
Cosine similarity adalah sebuah cara untuk mengukur seberapa mirip dua pengguna berdasarkan skor review yang mereka berikan untuk buku-buku.

Bagaimana Cara Kerjanya?


1.   Matriks Input: Kita masukkan pivot table ke dalam fungsi cosine similarity. Dalam konteks ini, setiap baris adalah pengguna, dan setiap kolom adalah judul buku.
2.   Menghitung Kemiripan: Fungsi cosine similarity akan menghitung kemiripan antara setiap pasangan pengguna. Hasilnya adalah matriks baru yang menunjukkan skor kemiripan antara pengguna-pengguna tersebut.
*   Skor 1 berarti pengguna sangat mirip (review buku yang mereka berikan sangat mirip).
*   Skor 0 berarti pengguna tidak mirip sama sekali.

Penggunaan dalam Sistem Rekomendasi
Kita bisa menggunakan matriks kemiripan ini untuk merekomendasikan buku kepada pengguna. Caranya:

1.   Cari Pengguna Paling Mirip: Temukan pengguna yang paling mirip dengan pengguna tertentu berdasarkan skor cosine similarity.
2.   Rekomendasikan Buku: Rekomendasikan buku-buku yang diberi nilai tinggi oleh pengguna yang mirip tadi, karena ada kemungkinan besar bahwa pengguna yang ingin kita rekomendasikan juga akan menyukai buku-buku tersebut.

```python
def recommend(book_name):
    index = np.where(pivot_table.index==book_name)[0][0]
    similar_books = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1], reverse=True)[1:6]

    data = []

    for i in similar_books:
        item = []
        temp_df = books[books['Title'] == pivot_table.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Title')['Title'].values))
        item.extend(list(temp_df.drop_duplicates('Title')['authors'].values))
        item.extend(list(temp_df.drop_duplicates('Title')['image'].values))

        data.append(item)
    return data
```

```python
recommend("Yoga: The Spirit and Practice of Moving into Stillness")
```

```python
from surprise import Dataset, Reader, SVD
from surprise.model_selection import train_test_split
from surprise import accuracy

# Define the rating scale
reader = Reader(rating_scale=(0, 10))

data = Dataset.load_from_df(df[['User_id', 'Title', 'review/score']], reader)
train_set, test_set = train_test_split(data, test_size=0.20, random_state=42)
model = SVD()
model.fit(train_set)
predictions = model.test(test_set)
accuracy.rmse(predictions)
```

```python
def recommend_books(user_id, n=10):
    # List all unique book titles
    all_books = df['Title'].unique()

    # Remove books already rated by the user
    rated_books = df[df['User_id'] == user_id]['Title'].values
    books_to_predict = [book for book in all_books if book not in rated_books]

    # Predict ratings for remaining books
    predictions = []
    for book in books_to_predict:
        pred = model.predict(user_id, book)
        predictions.append((book, pred.est))

    # Sort predictions by estimated rating
    predictions.sort(key=lambda x: x[1], reverse=True)

    # Get top N recommendations
    top_n = predictions[:n]

    return top_n
```

```python
user_id = 'A0461265TOJ3VFHJUMY7'
recommended_books = recommend_books(user_id)
print(f"Top 10 recommended books for user {user_id}:")
for i, (title, rating) in enumerate(recommended_books, start=1):
    print(f"{i}. {title} - Predicted Rating: {rating}")
```
